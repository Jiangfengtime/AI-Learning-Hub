函数调用

```python
from functools import reduce


def printInfo(name, age):
    print("姓名: ", name)
    print("年龄: ", age)
    print("-" * 20)

printInfo("张三", 20)
printInfo(age=20, name="李四")
```

默认值参数
1. 如果没传值, 则用默认值
2. 如果有传值, 则使用具体的值
3. 注意: 非默认值参数需要放在默认参数前

```python
def printInfo(name, age = 20):
    print("姓名: ", name)
    print("年龄: ", age)
    print("-" * 20)

printInfo("张三")
printInfo(age=18, name="李四")
```

不定长参数

加了星号 * 的参数会以元组(tuple)的形式导入，存放所有未命名的变量参数

```python
def printInfo(num, *tels):
    print(num)
    print(tels)
    print("-" * 20)

printInfo(1, 2, 3, 4, 5, 6)
```

还有一种就是参数带两个星号 **的可变长参数

加了两个星号 ** 的参数会以字典的形式导入，后面就不能再有其他参数了

```python
def printInfo(num, **tels):
    print(num)
    print(tels)
    print("-" * 20)

printInfo(5, name = "Tom", age = "18")
```

解包传参

```python
def func(a, b, c):
    return a + b + c

tuple1 = (1, 2, 3)
print(func(*tuple1))
print("-" * 20)

dict1 = {"a": 10, "b": 20, "c": 30}
print(func(**dict1))
```

当你在函数内部对一个变量进行赋值操作时，Python 默认会把这个变量当作局部变量，即便全局作用域中已经存在同名变量。

```python
# 当你在函数内部对一个变量进行赋值操作时，Python 默认会把这个变量当作局部变量，即便全局作用域中已经存在同名变量。
var1 = 100
def function_a():
    var1 += 200
print(var1) # 100
```

global: 将局部变量定义为全局变量

```python
def func():
    global var1
    var1 = 200
    print("var1: ", var1)

var1 = 100
print(var1)
func()
print(var1)
```

使用普通函数传参

```python
def operator(a, b):
    return a + b
def function(a, b, operator):
    return operator(a, b)

function(10, 20, operator=operator)
```

使用匿名函数传参

```python
def function(a, b, operator):
    return operator(a, b)

function(10, 20, lambda a, b: a + b)
```

匿名函数作为内置函数的参数
1. sorted:
2. map: map() 函数对序列中元素逐一处理。
3. filter: 函数对序列中元素过滤
4. reduce:

```python
student_list = [{"name": "张三", "age": "18"}, {"name": "李四", "age": "20"}, {"name": "王五", "age": "14"}]

print(student_list)
print(sorted(student_list, key=lambda x: x["age"]))
```

```python
# map() 函数对序列中元素逐一处理。
map_result = map(lambda x: x * x, [1,2, 3, 4])
print(list(map_result))
```

```python
# filter: 函数对序列中元素过滤
filter_result = filter(lambda x: x > 0, [1,2, -5, 0, 7])
print(list(filter_result))
```

```python
from functools import reduce
# 对元素进行累加
reduce_result1 = reduce(lambda x, y: x + y, [1,2,3,4,5])
print(reduce_result1)

reduce_result2 = reduce(lambda x, y: x * y, [1,2,3,4,5])
print(reduce_result2)
```
