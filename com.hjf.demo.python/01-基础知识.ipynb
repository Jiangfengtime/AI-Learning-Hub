1. type(): 获取变量类型

3. type()和isinstance()的区别在于: type()不会认为子类是一种父类类型; isinstance()会认为子类是一种父类类型

```python
# 整型
var1 = 100
var2 = True
var3 = True
print(f"type(var1) = {type(var1)}, type(var2) = {type(var2)}")
print(f"type(var1) = int, {type(var1) == int}")
print(f"isinstance(var1) = int ?  {isinstance(var1, int)}")
print(f"type(var2) = int, {type(var2) == int}") # False
print(f"isinstance(var2) = int ?  {isinstance(var2, int)}") # True, python中bool本质上是int的子类
```

```python
num1 = 21 # 10101
num2 = 15 # 01111
print(f"num1 = %d, num2 = %d, num1 & num2 = %d" % (num1, num2, num1 & num2)) # 101 = 5
print(f"num1 = %d, num2 = %d, num1 | num2 = %d" % (num1, num2, num1 | num2)) # 11111 = 31
print(f"num1 = %d, num2 = %d, num1 ^ num2 = %d" % (num1, num2, num1 ^ num2)) # 11010 = 26
print(f"num1 = %d, ~num1 = %d" % (num1, ~num1)) # -22
```

身份运算符

is/not is: 判断两个标识符是否引用自相同对象
1. python中 字符串创建时默认自动开启intern
2. python中 运行时创建的字符串默认不开启intern, 需手动开启

```python
import sys

m = 20
n = 20
q = 30
print(m is n)
print(m is q)
print(m is not n)
print(m is not q)
print("=" * 50)
str1 = "str"
str2 = "123"
str3 = str1 + str2
str4 = "".join(["str", "123"])
str5 = "str" + "123"
str6 = "str123"
str7 = sys.intern(str3)
str8 = sys.intern(str4)
print(str3 is str6) # False: 运行时创建的对象不会自动intern
print(str4 is str6) # False: 运行时创建的对象不会自动intern
print(str5 is str6) # True: 默认开启
print(str7 is str6) # True: 使用sys.intern手动开启intern
print(str7 is str6) # True: 使用sys.intern手动开启intern
```

### 流程分支

if - elif - else

```python
from random import randint

balance = randint(0, 100)
price = 50
print(f"余额: {balance}")
if balance < price:
    print("余额不足, 请充值")
else:
    balance -= price
    print(f"购买成功, 余额: {balance}")
print("欢迎光临")
```

match - case

```python
month = int(input("请输入月份: "))
# match month := 3: # 等价于: match 3
match month:
    case 1 | 3 | 5 | 7 | 8 | 10 | 12:
        print(f"{month}月有31天")
    case 4 | 6 | 9 | 11:
        print(f"{month}月有30天")
    case 2:
        print(f"{month}可能有28/29天")
    case _:
        print(f"{month} 月份错误")
```

三目运算符

表达式1 if 判断条件 else 表达式2

```python
num1, num2 = 3, 4
max_val = num1 if num1 > num2 else num2
print(f"最大值: {max_val}")
```

### 容器数据类型

列表、元组、字典和集合的区别
数据结构|是否可变|是否重复|是否有序|定义符号
--|--|--|--|--
列表(List)|可变|允许|有序|[]/list()
元组(Tuple)|不可变|允许|有序|()/tuple()
字典(Dictionary)|可变|键不允许,值允许|键无序|{}/dict()
集合(Set)|可变|不允许|无序|{}/set()




列表

```python
list1 = [100, 200, 300, 400, 500]
# 通过索引获取列表中的元素
print(list1[2]) # 300, 索引非负数, 从前往后数: [0, 1, 2, 3, 4]
print(list1[-3]) # 300, 索引负数, 从后往前数: [-5, -4, -3, -2, -1]
print("-" * 20)
# 列表切片
print(list1[:3]) # [100, 200, 300], 等价于: list1[0:3]
print(list1[2:4]) # [300, 400]
print(list1[::2]) # [100, 300, 500], 步长为2
print(list1[:-3]) # [100, 200], 等价于: list1[-5:-3]
print(list1[::-1]) # [500, 400, 300, 200, 100]
print("-" * 20)
```

列表推导式

```python
# 基础的列表推导式
squares = [x ** 2 for x in range(5)]
print("[x ** 2 for x in range(5)]: ", squares)

# 带条件的列表推导式
squares = [x ** 2 for x in range(10) if x % 2 == 0]
print("[x ** 2 for x in range(10) if x % 2 == 0]: ", squares)

# 使用现有列表的列表推导式
squares = [x ** 3 for x in [1, 2, 3, 4]]
print("[x ** 3 for x in [1, 2, 3, 4]]: ", squares)

# 包含多个循环的列表推导式
list1 = [1, 2, 3]
list2 = ["a", "b", "c"]
tuple_list = [(i, j) for i in list1 for j in list2]
print(tuple_list)
```
